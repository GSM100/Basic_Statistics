1. Types of Data

    -  Qualitative (Categorical) : Non-numeric, descriptive.

    -  Example: Gender (Male/Female), Colors, Blood type.

    -  Quantitative (Numerical): Numeric, measurable.

     -  Example: Age, Income, Height, Weight.

 Scales of Measurement :

  -  Nominal: Categories without order (e.g., blood group).

-  Ordinal: Ordered categories (e.g., survey rating: poor, good, excellent).

- Interval: Numeric scale without true zero (e.g., temperature in Celsius).

- Ratio: Numeric scale with true zero (e.g., height, weight, income).

2. Measures of Central Tendency

   -  Mean: Arithmetic average; used for symmetric distributions without outliers.

   -  Median: Middle value; used when data is skewed or has outliers (e.g., income data).

   -  Mode: Most frequent value; useful for categorical data (e.g., most common shoe size).

3. Dispersion

   -  Describes how spread out the data is.

   -  Variance: Average of squared deviations from mean.

    -  Standard Deviation (SD): Square root of variance; measures average spread in same units as data.

-  Higher SD → more spread; lower SD → more consistency.

4. Box Plot

    -  Visual summary of data distribution using 5-number summary: min, Q1, median,    -  Q3, max.

    -  Shows spread, skewness, and outliers.

5. Random Sampling

      -  Ensures each member of population has equal chance of being selected.

      -  Helps make unbiased inferences about population.

      -  Reduces selection bias and improves representativeness.

6. Skewness

    - Measure of asymmetry in data distribution.

    -  Positive skew (right-skewed): Tail on right, mean > median. Example: income.

     -  Negative skew (left-skewed): Tail on left, mean < median. Example: age at death in elderly population.

     -    Interpretation: Skewness affects which central tendency measure is more reliable.

7. Interquartile Range (IQR)

    - IQR = Q3 – Q1 (middle 50% of data).

   - Used to detect outliers:

   - Outlier if < Q1 – 1.5×IQR or > Q3 + 1.5×IQR.



8. Binomial Distribution Conditions

     - Fixed number of trials (n).

      -   Each trial has 2 possible outcomes (success/failure).

      - Probability of success (p) is constant.

      - Trials are independent.

      - Example: Flipping a coin 10 times, number of heads.



9. Normal Distribution & Empirical Rule

     - Symmetric, bell-shaped distribution.

     - Properties: mean = median = mode, total area = 1.

     - Empirical Rule:

     - 68% data within 1 SD

     - 95% within 2 SDs

     - 99.7% within 3 SDs

11. Random Variables

-  Random Variable: Function that assigns numerical values to outcomes of a random experiment.

- Discrete: Countable outcomes (e.g., number of heads in coin flips).

- Continuous: Infinite outcomes (e.g., height, weight).

12. Covariance & Correlation Example

- Dataset: Students’ hours studied (X) vs. exam score (Y).

- Covariance: Measures direction of relationship (positive/negative).

- Correlation: Standardized measure (–1 to +1) showing strength of relationship.

- +1 → perfect positive linear relation

- –1 → perfect negative linear relation

-  0 → no linear relation
